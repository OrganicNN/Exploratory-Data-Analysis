## Visualizing iris dataset 

```python
# Import libraries/packages
import pandas as pd
import seaborn as sns
sns.set(style="white", color_codes=True)
pd.set_option('display.max_colwidth', -1)
```

Iris data downloaded from:https://forge.scilab.org/index.php/p/rdataset/source/tree/master/csv/datasets/iris.csv)

This data has an unnamed column at the beginning which merely enumerates the rows in the dataset. Since Pandas enumerates for us, we don't want that row. Hence, we drop the unnamed column after reading it.

```python
# Read csv and get the column name to drop
iris = pd.read_csv('./Datasets/iris.csv')
iris.columns.values
```

```python
# Drop 'Unnamed: 0' column
iris = iris.drop(columns=['Unnamed: 0'])
iris.columns.values
```

Let us plot the iris dataset to visualize.

We know that there are multiple species. 

So let us see the differences between these species in terms of sepal length and width followed by petal length and width.

For both the visualizations, we will use seaborn's lmplot (http://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn-lmplot) 

```python
sns.lmplot('Sepal.Length', 'Sepal.Width', data=iris, hue='Species', fit_reg=False)
```

```python
sns.lmplot('Petal.Length', 'Petal.Width', data=iris, hue='Species', fit_reg=False)
```

In both the plots, we can see that Setosa is more distinctive than versicolor and virginica!
